```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
taxis = sns.load_dataset('taxis')
```

```python
taxis.head()
```

```python
taxis.info()
```

```python
sns.scatterplot(data=taxis,x="total",y="distance")
plt.title("Relationship between total and distance")
```

```python
sns.lineplot(x="distance", y="total",
             data=taxis)
```

```python
sns.scatterplot(x='fare', y="tip", hue="passengers", data=taxis)
plt.title('Fare vs Tip')
```

```python
sns.scatterplot(x='fare', y="tip", hue="color", data=taxis)
plt.title('Fare vs Tip')
```
